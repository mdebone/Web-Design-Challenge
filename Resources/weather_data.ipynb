```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
cities_data_df = pd.read_csv("cities.csv")
cities_data_df
```

```python
# Create a scatter plot which compares latitude to temperature
cities_data_df.plot(kind="scatter", x="Lat", y="Max Temp", grid=True, figsize=(8,8),
              title="City Latitude vs Max Temperature", color="#cc6600")
plt.savefig("lat_temp.png")
plt.show()
```

```python
# Create a scatter plot which compares latitude to humidity
cities_data_df.plot(kind="scatter", x="Lat", y="Humidity", grid=True, figsize=(8,8),
              title="City Latitude vs Humidity", color="#cc6600")
plt.savefig("lat_hum.png")
plt.show()
```
